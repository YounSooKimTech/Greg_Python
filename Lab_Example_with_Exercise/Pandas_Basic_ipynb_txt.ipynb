# crosstab

https://pandas.pydata.org/docs/reference/api/pandas.crosstab.html

```python
df
```

```python
import pandas as pd

pd.crosstab(index=df["gender"], columns=df["occupation"])
```

```python
pd.crosstab(index=df["gender"], columns=df["occupation"], values=df["age"], aggfunc = "mean") # aggfunc = aggreagated function
```

```python
pd.crosstab(index=df["gender"], columns=df["occupation"], values=df["age"], aggfunc = "mean").round(2) # aggfunc = aggreagated function
```

```python
# exercise:
"""
1. switch row and columns for the crosstab

2. compute the median age, according to the gender and occupation

"""

df3 = pd.crosstab(index=df["gender"], columns=df["occupation"], values=df["age"], aggfunc = "mean")
df3.transpose()

pd.crosstab(index=df["gender"], columns=df["occupation"], values=df["age"], aggfunc = "median")
```
