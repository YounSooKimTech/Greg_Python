# dummy variables

```python
import pandas as pd
import numpy as np

np.random.seed(123)

dog1_height = np.random.randint(6, 9, 10)
dog1_weight = np.random.randint(2, 6, 10)

dog_ch = pd.DataFrame({"height":dog1_height, "weight":dog1_weight})
dog_ch["kind"] = "chihuahua"

dog2_height = np.random.randint(21, 24, 10)
dog2_weight = np.random.randint(45, 60, 10)
dog_husky= pd.DataFrame({"height":dog2_height, "weight":dog2_weight})
dog_husky["kind"] = "husky"

print(dog_ch)
print(dog_husky)
```

```python
dogs = dog_ch.append(dog_husky)

dogs
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

model1 = smf.ols(formula="weight~height", data=dogs).fit()
print(model1.summary())

"""
This does not make sense. Look at the intercept. The weight starts from minus value.
Why? This is because we did not categorize the kind of dogs.
"""
```

```python
# excercise: convert the dogs["kind"] to dummy variable

import pandas as pd

dogs_dummies = pd.get_dummies(dogs.kind, prefix="kind")
dogs = pd.concat([dogs, dogs_dummies], axis=1)

dogs.head()



```

```python
model2 = smf.ols(formula = "weight~height+kind_chihuahua+kind_husky", data=dogs).fit()
print(model2.summary())

"""
Now, we have more realisitc result.
weight = 31 - 1.36*height -18*kind_chihuahua + 49*kind_husky

For instance, if I want to see the weight of chihuahua height=8,
weight = 31 - 1.36*8 -18*1 +49*0

In other way, it I want to figure out the weight of husky, height=22,
weight = 31 - 1.36*22 -18*0 + 49*1
"""
```

```python
model3 = smf.ols(formula = "weight~height+kind_husky", data=dogs).fit()
print(model3.summary())

"""
model3 is basically same as model2.
But in this case, I used chihuahua as base for dummy variable, kind.
In this case,
weight = 13.45 - 1.36*height + 67*d_kind

For instance, if I want to see the weight of chihuahua height=8,
weight = 31 - 1.36*8 +67*0 <- this is because base(0) is chihuahua

In other way, it I want to figure out the weight of husky, height=22,
weight = 31 - 1.36*22 +67*1 <- this is because d_kind==1 is husky

"""
```

```python
# 2. use mapping
# another way
dogs["d_kind"] = dogs["kind"].map({"chihuahua":0, "husky":1})
dogs.head()
```
