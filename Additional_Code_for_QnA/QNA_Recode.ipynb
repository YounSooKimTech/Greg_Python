```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/YounSooKimTech/Greg_Python/main/Data/GSS.2006.csv")
df.head()
```

```python
df = df[["socfrend", "big5a2"]]
df.info()
```

```python
conditions = [df["big5a2"]<3, df["big5a2"]>=3]
choices = [0,1]
df["sociable"] = np.select(conditions, choices)

df.head()

pd.crosstab(df["big5a2"], df["sociable"])
```

```python
conditions = [df["socfrend"]>=5, df["socfrend"]<5]
choices = [0,1]
df["high"] = np.select(conditions, choices)

df.head()

pd.crosstab(df["socfrend"], df["high"])
```
