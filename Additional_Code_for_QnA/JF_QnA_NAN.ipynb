```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/YounSooKimTech/Greg_Python/main/Data/GSS.2006.csv")
df.head()
```

```python
df1 = df.dropna(subset=["workhard"])
df1.head()
```

```python
df1.workhard.isnull().sum()
```

```python
conditions = [(df1.workhard < 3), (df1.workhard ==3), (df1.workhard >3)]
choices = ["realism", "neutral", "idealism"]

df1["status"] = np.select(conditions, choices)

df1.head()
```

```python
pd.crosstab(df1["workhard"], df1["status"])
```
