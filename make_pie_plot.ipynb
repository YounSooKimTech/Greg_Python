```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/YounSooKimTech/Greg_Python/main/GSS.2006.csv")
df.head()
```

```python

# 1 = white, 2=black, 3=others
race = df["race"].value_counts()
race = np.array(race)

plt.pie(race, labels=["white", "black", "others"], autopct='%.2f')
plt.title("race composition of GSS 2006")
plt.show()
```
